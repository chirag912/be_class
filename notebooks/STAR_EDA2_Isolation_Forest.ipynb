## Isolation forest

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
data = pd.read_csv("Iris.csv",index_col=0)
data.head()
```

```python
data_encoded = pd.get_dummies(data)
```

```python
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state = 10,contamination = .01)
clf.fit(data_encoded)
```

```python
#Predictions
y_pred_outliers = clf.predict(data_encoded)
```

```python
# -1 will stand for outliers and 1 for normal data
y_pred_outliers
```

```python
## Lets us add a new data point which is an obvious outlier
data_encoded.loc[150]=[20,40,30,50,1,0,0]
```

```python
data_encoded.tail()
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded['scores']=clf.decision_function(data_encoded)
```

```python
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:7])
```
