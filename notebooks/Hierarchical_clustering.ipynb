```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering 

# standard import libraries 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
univ = pd.read_csv("C:/Users/chira/Downloads/Universities.csv")
```

```python
univ
```

```python
# Normalization function

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(univ.iloc[:,1:])
```

```python
df_norm
```

```python
# create a dendogram 

dendrogram  = sch.dendrogram(sch.linkage(df_norm,method='centroid'))
```

```python
dendrogram  = sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
dendrogram  = sch.dendrogram(sch.average(df_norm))
```

```python
# crerate clusters

hc = AgglomerativeClustering(n_clusters = 3, affinity= 'euclidean', linkage = 'complete')
```

```python
# save and display  the clusters for the chart

y_hc = hc.fit_predict(df_norm)
Clusters= pd.DataFrame(y_hc,columns=['Clusters'])

Clusters
```

```python
df_norm['h_clusterid'] = Clusters
univ['h_clusterid'] = Clusters
```
