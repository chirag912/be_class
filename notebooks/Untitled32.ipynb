### Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv("/home/chirag/jupyter/be_class/Datasets/Iris.csv",index_col=0)
```

```python
data_encoded = pd.get_dummies(data)
```

```python
# train the model

clf = IsolationForest(random_state= 2, contamination = 0.02)
```

```python
clf.fit(data_encoded)
```

```python
y_pred_outliers  =clf.predict(data_encoded)
```

```python
y_pred_outliers
```
