```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
type(tips)
```

### Univariate  “one variable” data visualization

```python
#strip plot

sns.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
# Swarm plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
# Swarm plot

sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

### Box plot and Voilin plot 

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips')
```

```python
# Voilin plot

sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips')
```

### Combining Plots

```python
sns.violinplot(x='day',y='tip',data=tips,inner=None, color = 'lightgray' )
sns.stripplot(x='day',y='tip',data=tips, hue='sex')
plt.ylabel('tips($)')
plt.show()
```

### Bivariate plot data visualization 

```python
tips.head()
```

```python
# Joint plot 

sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
# Density plot

sns.jointplot(x='total_bill',y='tip', data=tips, kind='kde')
plt.show()
```

```python
# pairplot 

sns.pairplot(tips,hue='sex')
plt.show()
```
