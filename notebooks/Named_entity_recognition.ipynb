```python
noun_verbs = [token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(noun_verbs)
```

```python
import sklearn 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

X = cv.fit_transform(noun_verbs)
sum_words = X.sum(axis=0)
words_freq = [(word,sum_words[0,idx]) for word , idx in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x:x[1],reverse = True)
wf_df  =pd.DataFrame(words_freq)
wf_df.columns = ['word','count']
```

```python
wf_df[0:10]
```

```python
# Lets check the result in bar chart 

wf_df[0:10].plot.bar(x='word', figsize=(12,8), title='top verbs and nouns')
```

```python
import numpy as np 
import pandas as pd
import string
import spacy

from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import WordCloud
```

```python

emo_data = pd.read_csv("apple.txt",on_bad_lines='skip')
emo_data =  [x.strip() for x in emo_data.x]
emo_data = [x for x in emo_data if x]
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import tokenize
sentences = tokenize.sent_tokenize(" ".join(emo_data))
```

```python
sentences
```

```python
sent_df = pd.DataFrame(sentences,columns=['sentences'])
sent_df
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
```

```python
affinity_scores
```

```python

nlp = spacy.load('en_core_web_sm')
sentiment_lexicon = affinity_scores
```

```python
#custom fucntion : score each word in sentence 

nlp   = spacy.load('en_core_web_sm')
sentiment_lexicon =  affinity_scores

def calculate_sentiment(text: str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        print(sentence)
        for word in sentence:
            sent_score  = sent_score + sentiment_lexicon.get(word.lemma_,0)
    return sent_score
```

```python
calculate_sentiment(text = 'Ok movie')
```

```python
sent_df['sentiment_value'] = sent_df['sentences'].apply(calculate_sentiment)
```

```python
# how many words are there in the sentence

sent_df['word_count']= sent_df['sentences'].str.split().apply(len)
```
