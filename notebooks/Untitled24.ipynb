```python
# Normalization function
from sklearn.preprocessing  import StandardScaler
scaler = StandardScaler()

scaled_univ= scaler.fit_transform(univ.iloc[:,1:])

scaled_univ
```

```python
# Create the cluster algorithm 

from sklearn.cluster import KMeans
cluster_new = KMeans(3, random_state = 42)
cluster_new.fit(scaled_univ)
```

```python
# Assign the clusters to the OG dataset

univ['clusterid_new'] = cluster_new.labels_
```
